```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
import csv 
from sklearn import linear_model
import math


fdf=pd.read_csv('msample.csv')
fdf.head(10)
```

**Filling up Nan part with Median in case there is**

```python
Height=math.floor(fdf.Height.median())
len1=math.floor(fdf.Length1.median())
len2=math.floor(fdf.Length2.median())
len3=math.floor(fdf.Length3.median())
wid=math.floor(fdf.Width.median())

fdf.Height=fdf.Height.fillna(Height);
fdf.Width=fdf.Width.fillna(wid);
fdf.Length1=fdf.Length1.fillna(len1);
fdf.Length2=fdf.Length2.fillna(len2);
fdf.Length3=fdf.Length3.fillna(len3);

fdf.head(10)
```

```python
reg = linear_model.LinearRegression()
reg.fit(fdf[['Length1','Length2','Length3','Height','Width']], fdf.Weight)
```

```python
reg.coef_
```

```python
reg.intercept_
```

**Testing with tesing file to simplify the process***

```python
tdf=pd.read_csv('mtesting.csv')

tdf.head(10)
```

```python
tested=tdf.drop(columns='Weight')
tested.head(10)
```

```python
projection=reg.predict(tested)
print("prediction completed")
```

```python
projection
```

```python
tdf['Result']=projection
tdf
```

```python
%matplotlib inline 
plt.title("5 Different Fishes")
plt.xlabel('height')
plt.ylabel('weight')
plt.scatter(fdf.Height, fdf.Weight, color='red', marker='*')
plt.scatter(tdf.Height, tdf.Weight, color='green', marker='*')
```
